```python
'''
We solve this by making a set of seen and 
a set of tuples for output to store pairs
iterate through array and check for difference target
and then store pairs based on seen or not. 

'''
def array_pair_sum(arr, k):
	#first we do length check 
	if len(arr)<2:
		return "Array too short/No pairs"
	#if not then we make a set -> unique of seen 
	seen = set()
	#and an output set too 
	output = set()
	#start our loop to go through nums in array
	for num in arr: 
			target = k - num # difference logic 
			if target not in seen:
				seen.add(num)
			else: 
				output.add(min(num,target),max(num,target)) #a tuple for pair
	return output #or u could return len(output)
	
```

```python
#Implementing a stack 
'''
Stack is a LIFO data structure. 

'''
class Stack():
    def __init__(self):
        self.items = []
    def push(self,item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def isEmpty(self):
        return self.items == []
    def peek(self):
        return self.items[len(self.items)-1]
    def size(self):
        return len(self.items)
```

```python
# Implementing a queue 
'''
Queue is a FIFO data structure. 

'''
class Queue():
    def __init__(self):
        self.items = []
    def enqueue(self, item):
        self.items.insert(0,item) #insert at beginning of list to make it FIFO and pop from end
    def dequeue(self):
        self.items.pop()
    def isEmpty(self):
        return self.items == []
    def size(self):
        return len(self.items) 
```

```python
#Implementing a Deque 
'''
Deque is a double ended queue. 
#Add and remove from both ends 

'''
class Deque():
    def __init__(self):
        self.item = []
    def addFront(self,item):
        self.item.append(item)
    def addRear(self,item):
        self.item.insert(0,item)
    def removeFront(self):
        return self.item.pop()
    def removeRear(self):
        return self.item.pop(0)
    def isEmpty(self):
        return self.item == []
    def size(self):
        return len(self.item)
```

```python
'''
Balance check for paranthesis
s: '[]{}()[{()}]'
IF I DEFINE THESE FOR CHECKS 
open_stack = '({[' -> PUSH 
Match = {}()[]

'''
def balance_check(s):
  if len(s)%2 !=0:
    return False
  open = set(['(','[','{'])
  matches = set([('(',')'),('[',']'),('{','}')])
  stack = [] 
  for char in s: 
    if char in open: 
      stack.append(char)
    else: 
      if len(stack)==0:
        return False 
      last_open = stack.pop()
      if (last_open,char) not in matches:      
        return True
  return len(stack)==0     
        
```

```python
#implement queue using 2 stacks 
class Queue2Stacks(object):
    '''
    Implement a queue using 2 stacks 
    '''
  def __init__(self):
    self.stack1=[]
    self.stack2=[]

  def size(self):
    return len(self.stack1)+len(self.stack2)

  def enqueue(self, item):
    self.stack1.append(item)

  def dequeue(self):
    if len(self.stack2)==0:
      while(len(self.stack1)>0):
        self.stack2.append(self.stack1.pop())
    return self.stack2.pop()

  def isEmpty(self):
    return len(self.stack1)+len(self.stack2) == 0 
```

```python
#Checking cycle in a linked list 
def cycle_check(node):
    marker1 = node
    marker2 = node
    while marker2 != None and marker2.nextnode != None:
        marker1 = marker1.nextnode
        marker2 = marker2.nextnode.nextnode
        if marker1 == marker2:
            return True
    return False
```

```python
# reversing a linked list 
def reverse(head):
    current = head
    previous = None
    nextnode = None
    while current:
        nextnode = current.nextnode
        current.nextnode = previous
        previous = current
        current = nextnode
    return previous 
```

```python
# finding the nth node from the end of a linked list 
def nth_to_last_node(n,head):
    left_pointer = head
    right_pointer = head
    for i in range(n-1):
        if not right_pointer.nextnode:
            return None
        right_pointer = right_pointer.nextnode
    while right_pointer.nextnode:
        left_pointer = left_pointer.nextnode
        right_pointer = right_pointer.nextnode
    return left_pointer
```
