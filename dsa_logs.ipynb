```python
'''
We solve this by making a set of seen and 
a set of tuples for output to store pairs
iterate through array and check for difference target
and then store pairs based on seen or not. 

'''
def array_pair_sum(arr, k):
	#first we do length check 
	if len(arr)<2:
		return "Array too short/No pairs"
	#if not then we make a set -> unique of seen 
	seen = set()
	#and an output set too 
	output = set()
	#start our loop to go through nums in array
	for num in arr: 
			target = k - num # difference logic 
			if target not in seen:
				seen.add(num)
			else: 
				output.add(min(num,target),max(num,target)) #a tuple for pair
	return output #or u could return len(output)
	
```
