```python
def findMaxAverage(self, nums, k):
        curr_sum = sum(nums[:k]) #this is sum of all elements till k 
        max_sum = curr_sum

        # slide the window
        for right in range(k, len(nums)): #range starts from k to end of array
            curr_sum += nums[right] - nums[right - k] #this is basically the 
            #moving part where we add the new element and 
            #subtract the first from 
            #the old window [1,12,-5,-6,50]...... [12,-5-6,50]
            if curr_sum > max_sum:
                max_sum = curr_sum
        return max_sum / float (k) 
```
